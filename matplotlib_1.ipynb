```python
import matplotlib.pyplot as plt
x = ["python","c","c++", "java"]
y = [85,70,60,82]
plt.xlabel("language",fontsize=15)
plt.ylabel("number",fontsize=15)
plt.title("bar_plot_one",fontsize=15)
plt.bar(x,y,width=0.4,color="y")
plt.show()
```

```python
# alag alag color ke bars chahiye to
c=["y","b","m","g"]
plt.bar(x,y,width=0.4,color=c,align="center")
plt.show()
```

```python
#Highlighting the bar plot with red color
c=["y","b","m","g"]
plt.bar(x,y,width=0.4,color=c,edgecolor="r")
```

```python
plt.bar(x,y,width=0.4,color=c,edgecolor="r",linewidth=5)   #linewidth
```

```python
#Agar bar ke colors ko light karna ho to alpha parameter use
plt.bar(x,y,width=0.4,color=c,edgecolor="r",linewidth=5,alpha=0.5)
```

```python
z = [20,30,40,50]
width = 0.2
plt.bar(x,y,color="r",label="popularity")
plt.bar(x,z,color="y",label="popularity1")
plt.legend()
```

```python
import numpy as np
p = np.arange(len(x))
plt.bar(p,y,width,color="r",label="popularity")
plt.bar(p,z,width,color="y",label="popularity1")
plt.legend()
```

```python
# first x axis mey python c++ w likha hua aa rha tha ab 0, 1 2 aise numbers aa rhe hai
p1 = [j+width for j in p] # ye step ke baad 2 bar plot side by side ban jaayege
plt.bar(p,y,width=0.4,color="r",label="popularity")
plt.bar(p1,z,width=0.4,color="y",label="popularity1")
plt.legend()
```

#  horizontal_bar

```python
x = ["python","c","c++", "java"]
y = [85,70,60,82]
z = [20,30,40,50]

plt.xlabel("language",fontsize=15)
plt.ylabel("number",fontsize=15)
plt.title("horizontal_bar",fontsize=15)

plt.barh(x,y,color="r",label="popularity")
plt.barh(x,z,color="y",label="popularity1")
plt.legend()

plt.show()
```

```python
import numpy as np
p = np.arange(len(x))

plt.xlabel("language",fontsize=15)
plt.ylabel("number",fontsize=15)
plt.title("horizontal_bar",fontsize=15)

width=0.2
p1 = [j+width for j in p]
plt.barh(p,y,color="r",label="popularity")
plt.barh(p1,z,color="y",label="popularity1")
plt.legend()
plt.show()
```

# Scatter plot

```python
import matplotlib.pyplot as plt
day = [1,2,3,4,5,6,7]
number = [2,3,1,4,5,3,6]
plt.scatter(day,number,color="r")
plt.title("Scatter Plot",fontsize=15)
plt.xlabel("Day",fontsize=15)
plt.ylabel("Number",fontsize=15)
plt.show()
```

```python
colors = ["r","y","g","b","m","cyan","orange"]
plt.scatter(day,number,c=colors)
```

```python
colors = ["r","y","g","b","m","cyan","orange"]
sizes = [100,200,400,500,200,100,700] 
plt.scatter(day,number,c=colors,s=sizes)
```

```python
colors = ["r","y","g","b","m","cyan","orange"]
sizes = [100,200,400,500,200,100,700]
plt.scatter(day,number,c=colors,s=sizes,alpha=0.4)
plt.title("Scatter Plot",fontsize=15)
plt.xlabel("Day",fontsize=15)
plt.ylabel("Number",fontsize=15)
```

```python
colors = ["r","y","g","b","m","cyan","orange"]
sizes = [100,200,400,500,200,100,700]
plt.scatter(day,number,c=colors,s=sizes,marker="*")
plt.title("Scatter Plot",fontsize=15)
plt.xlabel("Day",fontsize=15)
plt.ylabel("Number",fontsize=15)
```

```python
colors = ["r","y","g","b","m","cyan","orange"]
sizes = [100,200,400,500,200,100,700]
plt.scatter(day,number,c=colors,s=sizes,marker="*",edgecolor="g",linewidth=4)
plt.title("Scatter Plot",fontsize=15)
plt.xlabel("Day",fontsize=15)
plt.ylabel("Number",fontsize=15)
```

```python
#  numbers ke basis color cmap parameter
colors = [10,49,30,29,56,20,30]
sizes = [100,200,400,500,200,100,700]
plt.scatter(day,number,c=colors,s=sizes,marker="*",cmap="Pastel1")
plt.title("Scatter Plot",fontsize=15)
plt.xlabel("Day",fontsize=15)
plt.ylabel("Number",fontsize=15)
```

 # Pie chart



```python
import matplotlib.pyplot as plt
x = [10,20,30,40]
y = ["C","C++","Java","Python"]

plt.pie(x,labels=y)

plt.show()
```

```python
#autopct -->value
ex = [0.3,0.0,0.0,0.0]
plt.pie(x,labels=y,explode=ex,autopct="%0.2f%%")
plt.show()
```

```python
#startangle
plt.pie(x,labels=y,startangle=180,autopct="%0.2f%%")
plt.show()
```

```python
#font ko badha chota karna ho to textprop parameter use karo and usme fontsize use karo
plt.pie(x,labels=y,startangle=0,autopct="%0.2f%%",textprops={"fontsize":20},radius=2) # overlap kam karne ke liye simply chart ka radius badha diya hai

plt.show()
```

```python
plt.pie(x,labels=y,autopct="%0.2f%%",counterclock=True)

plt.show()
```

```python
plt.pie(x,labels=y,autopct="%0.2f%%",counterclock=False)

plt.show()
```

```python
plt.pie([1])
plt.show()
```
